```python
from whu_supervision_scheduler import (
    WHUSupervisionScheduler
)

def remote_optimization_example():
    """Example optimizing for remote workers."""
    print("\n" + "=" * 60)
    print("Random OPTIMIZATION EXAMPLE - Minimize Campus Days")
    print("=" * 60)
    
    # Create scheduler optimized for remote workers
    scheduler = WHUSupervisionScheduler(
        weight_weeks=0,                 # High weight on minimizing weeks
        weight_days=0,                  # High weight on minimizing days
        weight_consecutive_weeks=0,     # Strongly prefer consecutive weeks
        weight_consecutive_days=0,      # Strongly prefer consecutive days
        weight_assignment_fairness=10,   # Still important but lower priority
        weight_minutes_fairness=10,      # Fairness in total minutes
        flex_discount=0                    # Larger discount for on-site chairs
    )
    scheduler.solver_timeout = 20
    
    scheduler.read_supervision_file('example/supervision_input.xlsx')
    
    if scheduler.schedule():
        print("\nRemote Worker Analysis:")
        print("-" * 40)
        
        # Analyze eco-friendly chair schedules
        for _, chair in scheduler.chairs_df.iterrows():
            if chair['Eco']:
                chair_name = chair['Chair']
                days = len(scheduler.solution['day_assignments'][chair_name])
                weeks = len(scheduler.solution['week_assignments'][chair_name])
                exams = scheduler.solution['chair_totals'][chair_name]
                
                print(f"{chair_name}: {exams} exams across {days} days in {weeks} weeks")
        
        scheduler.print_solution()
        scheduler.visualize_schedule()
        
        scheduler.write_solution_to_file('2_random_schedule_2.xlsx')

try:
    # Run remote optimization example
    remote_optimization_example()
    
except FileNotFoundError as e:
    print(f"\nError: {e}")
    print("Please ensure 'example/supervision_input.xlsx' exists in the current directory.")
except Exception as e:
    print(f"\nUnexpected error: {e}")
    import traceback
    traceback.print_exc()

```

```python
from whu_supervision_scheduler import (
    validate_excel_file
)

def validation_example():
    """Example showing input validation."""
    print("\n" + "=" * 60)
    print("VALIDATION EXAMPLE")
    print("=" * 60)
    
    # Test with a valid file
    print("Validating 'example/supervision_input.xlsx'...")
    is_valid, errors = validate_excel_file('example/supervision_input.xlsx')
    
    if is_valid:
        print("✓ File is valid!")
    else:
        print("✗ Validation errors found:")
        for error in errors:
            print(f"  - {error}")
            
    # Test with an invalid filename
    print("\nValidating 'nonexistent_file.xlsx'...")
    is_valid, errors = validate_excel_file('nonexistent_file.xlsx')
    
    if not is_valid:
        print("✗ Validation errors found:")
        for error in errors:
            print(f"  - {error}")

try:
    # Run validation example
    validation_example()
    
except FileNotFoundError as e:
    print(f"\nError: {e}")
    print("Please ensure 'example/supervision_input.xlsx' exists in the current directory.")
except Exception as e:
    print(f"\nUnexpected error: {e}")
    import traceback
    traceback.print_exc()

```
